```python
fig = px.bar(dataset, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
fig.show()
```

```python
fig.update_layout(title_text="Cars Listings by Brand", xaxis_title="Brand", yaxis_title="Cars Listings")
```

```python
fig.update_traces(marker=dict(color='red', opacity=0.7, line=dict(color='red', width=1.5)))
fig.update_layout(coloraxis_colorbar=dict(title="Cars Listings"))
```

```python
import plotly.express as px

df = px.data.gapminder().query("country=='United America'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in United America')
fig.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [100, 150, 200, 180, 250, 300]

# Create the area chart
plt.figure(figsize=(8, 6))
plt.fill_between(years, sales, color='skyblue', alpha=0.4)
plt.plot(years, sales, color='blue', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Trend of Global Sales Over the Years')
plt.grid(True)

# Display the chart
plt.show()
```
